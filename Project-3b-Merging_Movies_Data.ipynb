# Project - Cleaning, Merging and Transforming Movies Dataset

## OBJECTIVE

Importing clean movie data and unclean cast and crew data with help of pandas ,numpy, ast and json libraries and merge them according to their movie id which are in clean movie data and finally save data in csv format.

## Steps

1. Getting the Datasets from **movies_clean.csv** and **credits.csv**
2. Preparing the Data for Merge
3. Merging the Data from both dataframes
4. Cleaning and Transforming the new "Cast" Column
5. Cleaning and Transforming the new "Crew" Column
6. Drop Cast and Crew Columns
7. Save clean and merged movie data in a csv format
8. Reading cleaned movie data

### Deliverable

Save the clean and merged movie data in a csv format.

### Getting the Datasets

```python
import pandas as pd
```

```python
df = pd.read_csv("movies_clean.csv", parse_dates=["release_date"])
```

```python
df.info()
```

```python
credits = pd.read_csv("credits.csv")
```

```python
credits
```

```python
credits.info()
```

### Preparing the Data for Merge

```python
credits.id.value_counts()
```

```python
credits[credits.duplicated(subset = ["id"], keep = False)].sort_values("id")
```

```python
credits.drop_duplicates(subset = "id", inplace = True)
```

### Merging the Data

```python
df[~df.id.isin(credits.id)]
```

```python
credits[~credits.id.isin(df.id)]
```

```python
df = df.merge(credits, how = "left", left_on = "id", right_on = "id")
```

```python
df.head(10)
```

```python
df.info()
```

### Cleaning and Transforming the new "Cast" Column

```python
import ast
import numpy as np
```

```python
#df.cast[0]
```

```python
df.cast = df.cast.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
pd.DataFrame(df.cast[0])
```

```python
df["cast_size"] = df.cast.apply(lambda x: len(x) if isinstance(x,list) else np.nan)
```

```python
df.cast_size.value_counts(dropna = False).head(10)
```

```python
df.cast = df.cast.apply(lambda x: "|".join(i['name'] for i in x) if isinstance(x, list) else np.nan)
```

```python
df.cast
```

```python
df.cast.value_counts().head(10)
```

```python
df.cast.replace("", np.nan, inplace = True)
```

### Cleaning and Transforming the new "Crew" Column

```python
#df.crew[0]
```

```python
df.crew = df.crew.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
pd.DataFrame(df.crew[0])
```

```python
df["crew_size"] = df.crew.apply(lambda x: len(x) if isinstance(x,list) else np.nan)
```

```python
df.crew_size.value_counts(dropna = False).head(10)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df["director"] = df.crew.apply(lambda x:get_director(x) if isinstance(x,list) else np.nan)
```

```python
df.director
```

```python
df.director.value_counts(dropna = False).head(10)
```

### Drop Cast and Crew Columns

```python
df.info()
```

```python
df.drop(columns = ["cast","crew"], inplace = True)
```

### Save clean and merged movie data in a csv format

```python
df.to_csv("movies_complete.csv", index = False)
```
